## Подготовка данных:

```python
try:
  df = pd.read_excel('/datasets/for_project.xlsx', index_col=[0], parse_dates=[0])
except:
  df = pd.read_excel("/content/drive/MyDrive/For_data/for_project.xlsx", parse_dates=[0])
```

```python
df['time_on_foot'] = df['metrs']/ df['on_foot_speed_m_per_min']
```

```python
df['time_on_robo_car'] = df['metrs']/ df['on_robo_car_speed']
```

```python
save_time_per_day = pd.DataFrame({'n_day': [],
                       'time_on_foot': [], 'time_on_robo_car' : []})
```

```python
for i in range(365*5):
  save_time_per_day.loc[i, 'n_day']=i+1
  save_time_per_day.loc[i, 'time_on_foot'] = sum(df['time_on_foot']) *(i+1)
  save_time_per_day.loc[i, 'time_on_robo_car'] = sum(df['time_on_robo_car']) *(i+1)
```

```python
save_time_per_day['time_on_foot'] = save_time_per_day['time_on_foot'] /60
save_time_per_day['time_on_robo_car'] = save_time_per_day['time_on_robo_car'] /60
```

```python
save_time_per_day['saves_time'] = (save_time_per_day['time_on_foot'] - save_time_per_day['time_on_robo_car'])
```

```python
save_time_per_day['saves_time']
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=save_time_per_day['time_on_foot'], y=save_time_per_day['n_day'], name='On Feet'))
fig.add_trace(go.Scatter(x=save_time_per_day['time_on_robo_car'], y=save_time_per_day['n_day'], name='On Robo Car'))
fig.add_trace(go.Scatter(x=save_time_per_day['saves_time'], y=save_time_per_day['n_day'], name='saves_time'))
fig.show()
```
